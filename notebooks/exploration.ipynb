#### Non-nucleoside reverse-transcriptase inhibitors

```python
table1 = 'NNRTI_DataSet'
all_data = []
batch_size = 1000
offset = 0
```

```python
while True:
    resp = supabase.table(table1).select("*").range(offset, offset + batch_size - 1).execute()
    all_data.extend(resp.data)
    if len(resp.data) < batch_size:
        break
    offset += batch_size
```

```python
df1 = pl.DataFrame(all_data)
df1.shape
```
