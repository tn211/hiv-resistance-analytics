```python
from pathlib import Path
from math import ceil
import xml.etree.ElementTree as ET
```

```python
def list_xml_paths(xml_file, output_file):
    root = ET.parse(xml_file).getroot()
    paths = set()

    def walk(node, path=""):
        current = f"{path}/{node.tag}" if path else node.tag
        paths.add(current)
        for child in node:
            walk(child, current)

    walk(root)

    with open(output_file, "w") as f:
        for path in sorted(paths):
            f.write(path + "\n")

list_xml_paths("../data/xmls/xmls_db/1.xml", "paths.txt")
```

```python
with open("paths.txt") as f:
    data = [line.strip().removeprefix("tce/") for line in f if line.strip()]
    data = [d for d in data if d != "tce"]

# number of columns
n_cols = 3
rows = ceil(len(data) / n_cols)

# split into column chunks
cols = [data[i*rows:(i+1)*rows] for i in range(n_cols)]

# pad columns to equal length
for col in cols:
    while len(col) < rows:
        col.append("")

# compute max width per column
widths = [max(len(item) for item in col) for col in cols]

# print row-wise
for i in range(rows):
    row = ""
    for j in range(n_cols):
        row += cols[j][i].ljust(widths[j] + 4)
    print(row)
```
